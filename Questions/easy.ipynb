```python
#Write a function that returns the factorial of a number.

def Fact(n):
    if n<=1:
        return 1
    else:
        return n * Fact(n-1)
Fact(5)
```

```python
#Use map and lambda to convert a list of temperatures from Celsius to Fahrenheit.

cel=[20,30,40]
fer=(list(map(lambda x: (x*9/5)+32, cel)))
print(fer)
```

```python
#Use filter and lambda to get strings longer than 4 characters from a list.

x=["binal","yax","bhoomi"]
list(filter(lambda x: len(x)>4 , x))
```

```python
#Write a function that checks if a number is prime.

def Prime(num):
    if num<=1:
        return False
    for i in range(2, int(num*0.5)+1):
        if num%2==0:
            return False
    return True
    
print(Prime(6))
print(Prime(5))
```
