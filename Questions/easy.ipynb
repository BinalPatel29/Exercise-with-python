```python
#Write a function that returns the factorial of a number.

def Fact(n):
    if n<=1:
        return 1
    else:
        return n * Fact(n-1)
Fact(5)
```

```python
#Write a function that checks if a number is prime.

def Prime(num):
    if num<=1:
        return False
    for i in range(2, int(num*0.5)+1):
        if num%2==0:
            return False
    return True
    
print(Prime(6))
print(Prime(5))
```

```python
#A child class defines a method with the same name as a method in the parent class. Which method is called when using a child object?
class animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        return f"{self.name} Bark"
    
class Cat(animal):
    def speak(self):
        return f"{self.name} Meow"
    
A=animal("jambo")
print(A.speak())
B=Cat("billo")
print(B.speak())
```

```python
#A class prevents direct access to some of its attributes. How does this improve program reliability?
class Vote:
    def __init__(self, name, age):
        self.__name = name        # private
        self.__age = age          # private

    def set_age(self, age):
        if age >= 18:
            self.__age = age
            print(f"{self.__name} is eligible to vote")
        else:
            print(f"{self.__name} is not eligible to vote")

    def get_age(self):
        return self.__age

    def get_name(self):
        return self.__name

# Usage
a = Vote("Binal", 22)
print(a.get_name())
print(a.get_age())
a.set_age(17)       # demonstrates controlled access
```

```python
#A program uses a base class reference to call methods on different child class objects. What concept is being applied?
class animal:
    def __init__(self,name):
        self.name=name

class Dog(animal):
    def speak(self):
        return f"{self.name} says bark"

class Cat(animal):
    def speak(self):
        return f"{self.name} says meow"
    
animals=[Dog("Seru"),Cat("Billo")]
for a in animals:
    print(a.speak())
```

```python
#A class defines a method that must be implemented by all its child classes. What is the purpose of this design?
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def speak(self):
        pass  # must be implemented by all child classes

# Child classes
class Dog(Animal):
    def speak(self):
        return f"{self.name} says bark"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says meow"

# Objects of child classes
a = Dog("Seru")
b = Cat("Billo")

print(a.speak())  
print(b.speak())  
```

```python
#An object is compared using a comparison operator and returns a meaningful result instead of an error. How is this possible?
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Overload the '>' operator
    def __gt__(self, other):
        return self.age > other.age

    # Overload the '==' operator
    def __eq__(self, other):
        return self.age == other.age

# Objects
a = Animal("Billo", 5)
b = Animal("Chiku", 3)

# Comparison
print(a > b)   
print(a == b)  
```
