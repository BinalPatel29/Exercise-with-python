```python
# Tuple

# Question: Check whether a given value exists inside a tuple.

values=(22,33,44,55,66)
value=int(input("enter your value"))    #22
if value in values:
    print("Value exists")
else:
    print("Value doesn't exist")
```

```python
#Set

#Question:Create two sets and find how many elements are common between them.

a={1,2,3,4,5}
b={2,3,6,7,8}
print((a.intersection(b)))
print(len(b.intersection(a)))
```

```python
#Dictionary

#Question:Write a program to print all keys and values from a dictionary.

students={"Binal":90,"Bhoomi":70,"Riya":20}
print(students.items())

#or
for key,value in students.items():
    print(key,value)
```

```python
#Question:Write a program to check if a key exists in a dictionary.

students={"Binal":90,"Bhoomi":70,"Riya":20}
key=input("enter your key")   #Binal
if key in students:
    print("exists")
else:
    print("doesn't exist")
```
