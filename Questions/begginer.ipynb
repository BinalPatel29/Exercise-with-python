```python
#1. List:

# Question: Create a program that stores 5 numbers in a list and prints the largest number.

numbers = [12, 13, 35, 20, 29,90]
largest = numbers[0]
for number in numbers:
    if number > largest:
        largest = number
print(f"{largest} is bigger than all")
```

```python
#Use map to convert a list of integers [1,2,3,4] to their doubles.
numbers=[1,2,3,4]
list(map(lambda x: x*2, numbers))
```

```python
#Use filter to get only even numbers from a list [1,2,3,4,5,6].

number=[1,2,3,4,5,6]
list(filter(lambda x: x%2==0, number))
```
