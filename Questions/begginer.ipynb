```python
#A program creates a class and then creates multiple objects from it. What happens to the attributes for each object?
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"Name: {self.name}, age:{self.age}"
animal=Dog('seru',22)
print(animal)
```

```python
#A class contains a method that prints a message. An object calls this method. What is actually happening internally?
class animal:
    def dog(self):
        return "Bark"
    
A=animal()
print(A.dog())
```

```python
#A class has both instance variables and class variables. How are they stored and accessed?
class Dog:
    species="canine"      ## class variable
    def __init__(self, name, age):
        self.name=name     ## instance variable
        self.age=age       ## instance variable
    def __str__(self):
        return f"name: {self.name} , age= {self.age}"
B=Dog('Bittu',33)
print(B)
```

```python
#A child class is created from a parent class but has no additional methods or attributes. What behavior will the child class show?
class animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"name: {self.name} , age= {self.age}"
    
class dog(animal):
    pass

A=animal('diku',45)
B=dog('sweety',56)
print(A)
print(B)
```

```python
#An object’s data is accessed only through methods, not directly. Which OOP concept is this and why?
# encapsulation data
class Animal:
    def __init__(self, name, age):
        self.__name = name   # private variable
        self.__age = age     # private variable

    def get_name(self):
        return self.__name   # method to access name

    def get_age(self):
        return self.__age    # method to access age

A=Animal('chiku',47)
print(A.get_name())
print(A.get_age())
```

Exception handling & Advance python

```python
# A program uses a try block with a ZeroDivisionError except clause. The division succeeds.
try:
    a=5/1
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
```

```python
# An exception is raised inside a function but not handled there.
def divide():
    a = 5 / 0   # Exception raised inside the function

try:
    divide()
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# A loop iterates over a list using for x in obj.
numbers=[1,2,3,4,5]

for x in numbers:
    print(x)

# using iter
iterator=iter(numbers)
print(next(iterator))
print(next(iterator))
```

```python
# A function contains a yield statement and is called normally (without a loop).
# give lazy coding
n=[1,2,3,4,5]
def numbers(n):
    for number in n:
        yield number
a=numbers(n)
print(next(a))
print(next(a))
print(next(a))
```

```python
# A decorator prints a message before calling the original function.

# Decorator that prints a message before calling the function
# wrapper controls what happens before and after calling the original function, can also modify arguments or the return value
def announce(func):
    def wrapper(*args, **kwargs):
        print("About to call the function...")
        return func(*args, **kwargs)
    return wrapper

# Use the decorator
@announce
def car(name):
    print(f"My car is {name}")

# Call the decorated function
car("BMW")
```
