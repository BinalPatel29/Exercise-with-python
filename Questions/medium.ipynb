class and object

```python
#A child class accesses a protected attribute from its parent class. Why is this allowed but discouraged for external access?

class Animal:
    def __init__(self, name, age):
        self.name = name
        self._age = age

class Dog(Animal):
    def age(self):
        return f"Dog age is {self._age}"

# Objects
a = Dog("Billo", 5)
b = Dog("Chiku", 3)

print(a.age())
print(a._age)
print(b.age())
print(b._age)
```

```python
#A program replaces a parent class object with a child class object without breaking functionality. Why does this work?

class Animal:
    def speak(self):
        return "Animal makes a sound"
    
    def __init__(self,name,age):
        self.name=name
        self.age=age

class Dog(Animal):
    def speak(self):
        return "Dog barks"

A=Animal("Sheru",35)
print(A.name)
print(A.age)
print(A.speak())

B=Dog("Billo",55)
print(B.name)
print(B.age)
print(B.speak())
```

```python
#A class hides complex internal logic but exposes simple methods for interaction. How does this help developers?
class Calculator:
    def add(self, a, b):
        # Internal logic is hidden
        return a + b

# Developer uses only the simple method
calc = Calculator()
print(calc.add(5, 3)) 
```

```python
#Two unrelated classes respond correctly to the same operation. What OOP principle does this illustrate?
#polymorphism
class Dog:
    def speak(self):
        return "Dog barks"

class Cat:
    def speak(self):
        return "Cat meows"

# Both classes respond to the same operation
animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())
```

```python
#An object is printed and displays meaningful information instead of a memory address. What enables this behavior?
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"

# Create object
a = Animal("Billo", 5)
print(a)
```

```python
# A generator raises an exception after yielding some values.
def numbers():
    yield 1
    yield 2
    raise ValueError("somthing went wrong!")
    yield 3    # this is never reached

a=numbers()
print(next(a))
print(next(a))
print(next(a))
```

```python
# An iterator resets its internal counter inside __iter__() instead of __next__().
class GoodCounter:
    def __init__(self, data):
        self.data = data
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration
        val = self.data[self.index]
        self.index += 1
        return val

# Using the correct iterator
numbers = GoodCounter([1, 2, 3, 4, 5])

for x in numbers:
    print(x)

print("Next loop:")

# Second loop does not produce values
for x in numbers:
    print(x)
```

```python
# A decorator accepts arguments and decorates a function.
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def greet():
    print("Hello!")

greet()
```

```python
# A generator expression is used instead of a list comprehension inside a function.
def squares(n):
    return (x**2 for x in range(n))  # generator expression

for val in squares(5):
    print(val)
```

```python
# Create a figure with two subplots: one line plot and one bar chart.
import matplotlib.pyplot as plt
import numpy as np
x=["B.tech","C.E.","Civil","Mechanical"]
y1=[22,23,24,25]
y2=[3,4,5,6]
fig,axes=plt.subplots(1,2,figsize=(12,5))

#Bar chart
axes[0].bar(x,y1,color='skyblue')
axes[0].set_title("Bar Chart")
axes[0].set_xlabel("Course")
axes[0].set_ylabel("Value")

#Line chart
axes[1].plot(x,y2,marker='o',color='red',linestyle='-')
axes[1].set_title("Line Chart")
axes[1].set_xlabel("Course")
axes[1].set_ylabel("Value")

plt.tight_layout()
plt.show()
```

fig  the whole figure (canvas)

1, 2  1 row  2 columns of subplots (side by side)
figsize=(12, 5)  size of the figure in inches

```python
# Create a box plot to compare the distribution of a numeric variable across different categories.
import pandas as pd
df=pd.read_csv('/workspaces/Exercise-with-python/Employee.csv')

# Create a box plot: Age by Education
plt.figure(figsize=(10, 6))
df.boxplot(column='Age', by='Education', grid=True, patch_artist=True)

# Add titles and labels
plt.title("Distribution of Age by Education")
plt.suptitle("")  # Removes the automatic 'Boxplot grouped by Education' title
plt.xlabel("Education")
plt.ylabel("Age")

# Show the plot
plt.show()
```

```python
# Identify missing values in a DataFrame and replace them with the column mean.
print(df.isnull().sum())

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(numeric_cols)

for col in numeric_cols:
     df[col] = df[col].fillna(df[col].mean())

print("\nAfter filling missing values:")
print(df.isnull().sum())
```
