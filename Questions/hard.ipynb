```python
#Write a function that takes a list of numbers and returns a dictionary with number: square.

def Square(number):
    re={}
    for n in number:
        re[n]=n**2
    return re
number=[1,2,3,4]
print(Square(number))
```

```python
#Use map and filter together: double the even numbers from a list and ignore the odd numbers.

number=[2,3,4,5,6]
y=list(filter(lambda x: x%2==0,number))
print(y)
z=list(map(lambda w: w*2,y))
print(z)

#or
print(list(map(lambda x: x*2, filter(lambda x: x%2==0, number))))
```

```python
#Write a program that reads all text files from a directory and merges them into a single file.

import os

try:
    with open("merged.txt", "w") as outfile:
        for file in os.listdir("my_folder"):
            if file.endswith(".txt"):
                with open(os.path.join("my_folder", file), "r") as f:
                    outfile.write(f.read() + "\n")
except Exception as e:
    print("Error:", e)
```

```python
#Write a program that uses a standard library module to search for a file inside a directory and its subdirectories.

import os

directory = input("Enter the directory to search: ")
filename_to_find = input("Enter the file name to search for: ")

found = False

for root, dirs, files in os.walk(directory):
    if filename_to_find in files:
        print("File found at:", os.path.join(root, filename_to_find))
        found = True
        break  # remove this if you want to find all occurrences

if not found:
    print("File not found.")
```

```python
#Write a program that reads a log file and extracts only the lines containing errors into a new file.
# File paths
input_file = "log.txt"       # your log file
output_file = "errors.txt"   # file to save error lines

try:
    with open(input_file, "r") as infile, open(output_file, "w") as outfile:
        for line in infile:
            if "error" in line.lower():   # case-insensitive search
                outfile.write(line)
    print("Error lines extracted successfully.")
except FileNotFoundError:
    print(f"Error: The file '{input_file}' does not exist.")
except IOError:
    print("Error: An error occurred while reading or writing the file.")
```

```python
#Write a program that imports multiple standard library modules and uses them together to perform a file-based task.

import os
import glob
from datetime import datetime

# Directory containing text files
source_dir = "my_folder"
output_file = "report.txt"

# Get all text files in the directory
text_files = glob.glob(os.path.join(source_dir, "*.txt"))

with open(output_file, "w") as report:
    report.write(f"Word Count Report - {datetime.now()}\n")
    report.write("="*40 + "\n")  
    for file_path in text_files:
        try:
            with open(file_path, "r") as f:
                content = f.read()
                word_count = len(content.split())
            report.write(f"{os.path.basename(file_path)}: {word_count} words\n")
        except Exception as e:
            report.write(f"{os.path.basename(file_path)}: Error reading file ({e})\n")
print(f"Report generated successfully in '{output_file}'")
```

```python
#A deep inheritance hierarchy becomes difficult to maintain. What OOP design issue does this indicate?
class LivingBeing:
    pass

class Animal(LivingBeing):
    pass

class Mammal(Animal):
    pass

class Dog(Mammal):
    pass

class Puppy(Dog):
    pass

# Deep hierarchy: Puppy -> Dog -> Mammal -> Animal -> LivingBeing
p = Puppy()
print(isinstance(p, LivingBeing))  
```

```python
#A class depends only on abstract behavior rather than concrete implementations. Why is this beneficial?

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Works with any Animal
def make_speak(animal: Animal):
    print(animal.speak())

make_speak(Dog())
```

```python
#Two objects of different classes are added together using an operator. How does the program know what to do?
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overload the + operator
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Objects of the same class
p1 = Point(2, 3)
p2 = Point(4, 5)

# Adding objects
p3 = p1 + p2  # Calls p1.__add__(p2)
print(p3)
```

```python
#A class restricts how its attributes can be modified but still allows reading them. What balance is achieved here?
#encapsulation
class BankAccount:
    def __init__(self, balance):
        self._balance = balance  # "protected" internal attribute

    # Allow reading
    @property
    def balance(self):
        return self._balance

# Usage
account = BankAccount(1000)
print(account.balance)  # Allowed → 1000
account.balance = 500   # Error! Cannot modify directly
```

```python
#A system uses polymorphism heavily instead of conditional logic. Why is this considered better design?
class Animal:
    def speak(self):
        pass  # Interface method

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Polymorphism in action
animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())
```
