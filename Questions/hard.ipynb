```python
#A deep inheritance hierarchy becomes difficult to maintain. What OOP design issue does this indicate?
class LivingBeing:
    pass

class Animal(LivingBeing):
    pass

class Mammal(Animal):
    pass

class Dog(Mammal):
    pass

class Puppy(Dog):
    pass

# Deep hierarchy: Puppy -> Dog -> Mammal -> Animal -> LivingBeing
p = Puppy()
print(isinstance(p, LivingBeing))  
```

```python
#A class depends only on abstract behavior rather than concrete implementations. Why is this beneficial?

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Works with any Animal
def make_speak(animal: Animal):
    print(animal.speak())

make_speak(Dog())
```

```python
#Two objects of different classes are added together using an operator. How does the program know what to do?
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overload the + operator
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Objects of the same class
p1 = Point(2, 3)
p2 = Point(4, 5)

# Adding objects
p3 = p1 + p2  # Calls p1.__add__(p2)
print(p3)
```

```python
#A class restricts how its attributes can be modified but still allows reading them. What balance is achieved here?
#encapsulation
class BankAccount:
    def __init__(self, balance):
        self._balance = balance  # "protected" internal attribute

    # Allow reading
    @property
    def balance(self):
        return self._balance

# Usage
account = BankAccount(1000)
print(account.balance)  # Allowed → 1000
account.balance = 500   # Error! Cannot modify directly
```

```python
#A system uses polymorphism heavily instead of conditional logic. Why is this considered better design?
class Animal:
    def speak(self):
        pass  # Interface method

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Polymorphism in action
animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())
```

Exception handling & Advance python

```python
# A decorator wraps a generator function and logs when it starts and ends.
def log_gen(func):
    def wrapper(*args, **kwargs):
        print(f"{func.__name__} started")
        yield from func(*args, **kwargs)   # short way to yield all values
        print(f"{func.__name__} ended")
    return wrapper

@log_gen
def my_gen(n):
    for i in range(n):
        yield i

for val in my_gen(5):
    print(val)
```

```python
# An iterator implements both __iter__() and __next__() but returns a new object from __iter__().
class MyIterator:
    def __init__(self, n):
        self.n = n

    def __iter__(self):
        # return a fresh iterator object each time
        return MyIteratorIter(self.n)

class MyIteratorIter:
    def __init__(self, n):
        self.n = n
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.n:
            val = self.current
            self.current += 1
            return val
        else:
            raise StopIteration

# Using the iterator
for x in MyIterator(5):
    print(x)
```

```python
# An exception is raised inside a decorator before calling the wrapped function.
def check_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("x must be positive!")
        return func(x)
    return wrapper

@check_positive
def square_root(x):
    return x ** 0.5

# Example usage
print(square_root(9))   # works
print(square_root(-5))  # raises ValueError#
```

```python
# A generator uses try-except-finally and is only partially consumed.
def my_gen(n):
    try:
        for i in range(n):
            yield i
    except GeneratorExit:
        print("Generator was closed early!")
    finally:
        print("Cleanup in finally block")

# Create generator
gen = my_gen(5)

# Partially consume it
print(next(gen))  # 0
print(next(gen))  # 1

# Close generator early
gen.close()
```

```python
# Multiple decorators are stacked on a function, one of which handles exceptions.
def log(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

def handle_exceptions(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"Error: {e}")
    return wrapper

@log
@handle_exceptions
def divide(a, b):
    return a / b

# Examples
print(divide(10, 2))  # works
print(divide(5, 0))   # handled by exception decorator
```

```python
#Write Python code where multiple threads safely insert records into the same SQLite database.
import sqlite3
import threading

# Lock to synchronize database access
db_lock = threading.Lock()

# Function for threads to insert a user
def insert_user(name, age):
    with db_lock:  # ensure only one thread writes at a time
        conn = sqlite3.connect('app.db', check_same_thread=False)
        cursor = conn.cursor()
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS users (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                name TEXT,
                age INTEGER
            )
        """)
        cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", (name, age))
        conn.commit()
        conn.close()
        print(f"Inserted user: {name}, age: {age}")

# List of users to insert
users = [("Alice", 25), ("Bob", 30), ("Charlie", 22)]

# Create and start threads
threads = [threading.Thread(target=insert_user, args=user) for user in users]
for t in threads:
    t.start()
for t in threads:
    t.join()

print("All users inserted successfully.")
```

```python
#Write Python code using a context manager (with) to handle SQLite transactions with commit and rollback.
import sqlite3
import logging

# Configure logging
logging.basicConfig(
    filename='app.log',
    filemode='a',
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.INFO
)

user_name = "Alice"
user_age = 30

try:
    with sqlite3.connect('app.db') as conn:  # context manager handles commit/rollback
        cursor = conn.cursor()
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS users (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                name TEXT NOT NULL,
                age INTEGER
            )
        """)
        cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", (user_name, user_age))
    logging.info(f"User '{user_name}' inserted successfully.")

except sqlite3.Error as e:
    logging.error(f"Failed to insert user '{user_name}'. Error: {e}")
```

```python
#Write Python code that implements rotating log files using the logging module.
import logging
from logging.handlers import RotatingFileHandler

# Create a rotating file handler
handler = RotatingFileHandler(
    "app.log",      # Log file name
    maxBytes=1024,  # Max size in bytes before rotation
    backupCount=3   # Number of backup files to keep
)

# Configure logging
logging.basicConfig(
    handlers=[handler],
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.INFO
)

# Example logging
for i in range(100):
    logging.info(f"Logging message {i+1}")

print("Logging with rotation completed.")
```

```python
#Write Python code that runs a CPU-bound task using both threading and multiprocessing.
import threading
import multiprocessing
import time

# CPU-bound task: calculate sum of squares
def cpu_task(n):
    total = 0
    for i in range(n):
        total += i * i
    return total

N = 10_000_000

# ----------------- Using Threads -----------------
start = time.time()
threads = [threading.Thread(target=cpu_task, args=(N,)) for _ in range(4)]
for t in threads: t.start()
for t in threads: t.join()
print("Threading time:", time.time() - start)

# ----------------- Using Processes -----------------
start = time.time()
processes = [multiprocessing.Process(target=cpu_task, args=(N,)) for _ in range(4)]
for p in processes: p.start()
for p in processes: p.join()
print("Multiprocessing time:", time.time() - start)
```
