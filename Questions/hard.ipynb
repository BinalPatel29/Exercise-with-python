```python
# Real-World Problem

#Question: Write a program to store product names and prices in a dictionary and find the most expensive product.

products={"laptop":70000,"tv":50000,"fridge":180000,"machine":60000}
print(max(products,key=products.get),max(products.values()))

#or
most_expensive = max(products, key=products.get)
price = products[most_expensive]
print(f"The most expensive product is {most_expensive} with price {price}")
```

```python
#Write a function that takes a list of strings and returns the longest string.
def largest(str):
    return max(str, key=len)
str=["binal","bhoomi"]
print(largest(str))
```

```python
#Use map and lambda to convert a list of strings representing numbers into integers and then double them.

number=["3","4","5","6"]
z=list(map(lambda x: int(str(x)),number))
print(z)
w=list(map(lambda y : y*2 ,z))
print(w)
```

```python
#Use filter and lambda to find all prime numbers in a list.

number=[21,22,23,24,25]
list(filter(lambda x: x>1 and all( x % i !=0 for i in range(2, int(x**0.5)+1)),number))
```

```python
#Write a function that takes a list of numbers and returns a dictionary with number: square.

def Square(number):
    re={}
    for n in number:
        re[n]=n**2
    return re
number=[1,2,3,4]
print(Square(number))
```

```python
#Use map and filter together: double the even numbers from a list and ignore the odd numbers.

number=[2,3,4,5,6]
y=list(filter(lambda x: x%2==0,number))
print(y)
z=list(map(lambda w: w*2,y))
print(z)

#or
print(list(map(lambda x: x*2, filter(lambda x: x%2==0, number))))
```

```python
#Write a program that reads all text files from a directory and merges them into a single file.

import os

try:
    with open("merged.txt", "w") as outfile:
        for file in os.listdir("my_folder"):
            if file.endswith(".txt"):
                with open(os.path.join("my_folder", file), "r") as f:
                    outfile.write(f.read() + "\n")
except Exception as e:
    print("Error:", e)
```

```python
#Write a program that uses a standard library module to search for a file inside a directory and its subdirectories.

import os

directory = input("Enter the directory to search: ")
filename_to_find = input("Enter the file name to search for: ")

found = False

for root, dirs, files in os.walk(directory):
    if filename_to_find in files:
        print("File found at:", os.path.join(root, filename_to_find))
        found = True
        break  # remove this if you want to find all occurrences

if not found:
    print("File not found.")
```

```python
#Write a program that reads a log file and extracts only the lines containing errors into a new file.
# File paths
input_file = "log.txt"       # your log file
output_file = "errors.txt"   # file to save error lines

try:
    with open(input_file, "r") as infile, open(output_file, "w") as outfile:
        for line in infile:
            if "error" in line.lower():   # case-insensitive search
                outfile.write(line)
    print("Error lines extracted successfully.")
except FileNotFoundError:
    print(f"Error: The file '{input_file}' does not exist.")
except IOError:
    print("Error: An error occurred while reading or writing the file.")
```

```python
#Write a program that imports multiple standard library modules and uses them together to perform a file-based task.

import os
import glob
from datetime import datetime

# Directory containing text files
source_dir = "my_folder"
output_file = "report.txt"

# Get all text files in the directory
text_files = glob.glob(os.path.join(source_dir, "*.txt"))

with open(output_file, "w") as report:
    report.write(f"Word Count Report - {datetime.now()}\n")
    report.write("="*40 + "\n")  
    for file_path in text_files:
        try:
            with open(file_path, "r") as f:
                content = f.read()
                word_count = len(content.split())
            report.write(f"{os.path.basename(file_path)}: {word_count} words\n")
        except Exception as e:
            report.write(f"{os.path.basename(file_path)}: Error reading file ({e})\n")
print(f"Report generated successfully in '{output_file}'")
```
